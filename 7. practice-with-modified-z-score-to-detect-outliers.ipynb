# Practice with Modified Z Score to detect outliers

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/kaggle/input/modified-zscore-to-detect-outliers-practice/movie_revenues.csv')
```

```python
df
```

```python
# Checking statistics revenue collumn

df['revenue'].describe()
```

```python
# Create a new revenue collumn and convert into million of dollars

df['revenue_mln'] = df['revenue'].apply(lambda x: x/1000000)
```

```python
# Checking statistics of new revenue_mln collumn

df['revenue_mln'].describe()
```

```python
# Taking mean and std as variables from revenue_mln collumn
# by skipping first (_,) and the last ones (*_)

_, mean, std, *_ = df.revenue_mln.describe()
```

```python
mean
```

```python
std
```

```python
# Creating a function to calculate z score

def get_z_score(value, mean, std):
    return (value - mean)/std
```

```python
# Creating a Z Score collumn applied to revenue_mln collumn

df['z_score'] = df['revenue_mln'].apply(lambda x: get_z_score(x, mean, std))
```

```python
df.head()
```

```python
# Creating a function to calculate MAD

def get_mad(s):
    median = np.median(s)
    diff = abs(s - median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(df['revenue_mln'])
median = np.median(df['revenue_mln'])
```

```python
MAD
```

```python
median
```

```python
# Creating a function to calculate modified Z score

def get_modified_z_score(x, media, MAD):
    return 0.6745 * (x - median) / MAD
```

```python
# Creating a modified Z Score collumn applied to revenue_mln collumn

df['mod_z_score'] = df['revenue_mln'].apply(lambda x: get_modified_z_score(x, median, MAD))
```
